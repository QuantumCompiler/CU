* How do you know that `2+2` is equal to `4`?
* How do you know that lettuce expression `let x = 10 in x` and javascript expression `var x = 10 ; x;` and bash expression `x=10 ; x;` and scala expression `val x = 10 ; x;` all equal `10`?
* These are easy expressions to know the value of intuitivly, but there is logic behind your knowledge.
* While out of scope for 3155, I often find that it is helpful to understand the judgment derivations over sets operational semantics
* Derivations are not proofs of correctness or completeness for our language
* Derivations demonstrate that for some valid expression in the langauge, we are able find a value of that expression
* We can demonstrate that `2 + 2` evaluates to `4`
* We can demonstrate that `let x = 10 in x` evaluates to `10`
* We can demonstrate that `let a = 20 in let b = (c) { c + a } in let d = (e) { let a = 30 in b(e) } in d(40)` evaluates to `60` under static scoping conditions
    * and that same expression evaluates to `70` under dynamic scoping conditions
    * and taht same expression can have other values, or even errors under other kinds of logical expressivness defined by our operational semantics


## Definitions
* Judgment Form: a format to define some operation
    * e.g. eval(e) = v is a judgment form
    * e.g. v + v = v is a judgment form
* Judgment: a fact that can be demonstrated by preforming a derivation over complete sets of operational semantics over various judgment forms
    * e.g. eval(Const(5)) = 5 is a judgment
    * e.g. eval(Plus(Const(5), Const(10))) = 15 is a judgment
* Opeartional Semantic Rules: a rule used to define how one evaluates some judgment form (the thing with stuff above and below the line)
    * Conclusion: in the context of an operational semantic rule, the thing below the line that we want to assert as true
    * Premise: in the context of an operational semantic rule, the things above the line which must hold true for the conclusion to be derived
    * Axiom: an operational semantic rule with nothing above the line, no premis, only a conclusion
* Derivations:
  * Given some set of operational semantic rules
  * and given a sample program
  * demonstrate with logic what the value of that program is

# Adding Environments
* Let's extend our langauge with immutable variables, we'll need environments for that. 
* We'll use the judgment form from class `eval(e,env) = v`
* Let's define the env
    * {} is the empty environment
    * {... id -> v, ...} is a non-empty environment
    * {...} is an environment that may or may not be empty
    * {...} + id -> v will extend the environment with an id and value (v) pair
    (
* Judgment forms: 
    * {...} + id -> v = {...}
       * extend an environment with an id-value pair to get some new environment
    * {...} at id = v
       * get the value in the environment mapped to our id
    * id in {...} = b
        * true or false, the id is in this envrionment
    * id not in {...} = b
        * true or false, the id is in this envrionment
* Operationaly semantics for the enviornment
    * {} + id -> v = {id -> v}  >>> EXTEND-EMPTY-ENV
    * {..., id_i -> v_i, ... } + id_i -> v_i_p = {..., id_i -> v_i_p, ...}  >>> UPDATE-ENV
    * {...} + id -> v = {..., id -> v}  >>> EXTEND-NON-EMPTY-ENV
        * id not in {...}
    * {...} at id = error  >>> AT-ERROR
        * id not in {...}
    * {..., id -> v, ...} at id = v  >>> AT-OK
    * id in {..., id -> v, ... } = true  >>> IN-TRUE
    * id in {id_1 -> v1, ..., id_n -> v_n} = false  >>> IN-FALSE
        * id_1 != id
        * ...
        * id_n != id
        * *for all n 1 to n*
    * id not in {...} = bp  >>> NOT-IN
        * id in {...} = b
        * !b = bp

* Finally we have our operational semantics. I'll keep it small with CONST, PLUS, LET, and IDENT. We'll also ignore errors for this
* judgment form: `eval(e,{...}) = v`
* operational semantics:
    * eval(Const(v),{...}1) = v  >>> CONST-OK
    * eval(Plus(e1,e2), {...}1) = v3  >>> PLUS-OK
        * eval(e1, {...}1) = v1
        * eval(e2, {...}1) = v2
        * v1 + v2 = v3
    * eval(Ident(id), {...}1) = v   >>> IDENT-OK
        * id in {...}1 = true
        * {...}1 at id = v
    * NOTE: there would be an error case to write for ident, but I'm not dealing with errors for this writeup
    * eval(Let(id,e1,e2), {...}1) = v2   >>> LET-OK
        * eval(e1, {...}1) = v1
        * {...}1 + id -> v1 = {...}2
        * eval(e2, {...}2) = v2
